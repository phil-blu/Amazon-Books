#  EXPLORATORY DATA ANALYSIS

OBJECTIVE: TO DETERMINE WHAT CATEGORY OF BOOK/S ON AMAZON CUSTOMERS ARE MOST INTERSTED IN.

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import os
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
pd.set_option("display.max_rows", None)
```

```python
books = pd.read_csv("Most_Popular_Amazon_Books").drop("Unnamed: 0", axis=1)
books.head()
```

```python
books = pd.DataFrame(books.groupby(["CATEGORY", "TITLE", "AUTHOR", "DATE", "RATING"])["STAR"].mean()).reset_index()
books.head(10)
```

```python
most_purchased = books.sort_values(by=["RATING", "STAR"], ascending=False)
most_purchased
```

```python
if not os.path.exists("images"):
    os.mkdir("images")
#sample of most purchased books by category
import plotly.figure_factory as ff

df = most_purchased
df_sample = df[:6]

fig =  ff.create_table(df_sample)
fig.show()
fig.write_image("images/Most_purchased_books.jpeg")
```

```python
books["RATING"].max()
```

```python
df = books
fig = px.bar(df, y='RATING', x='CATEGORY', text='RATING')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.update_layout(title="NUMBER OF RATINGS FOR EACH CATEGORY")
fig.show()
fig.write_image("images/ratings.jpeg")
```

```python
recent = books.sort_values("DATE", ascending = False).drop(["CATEGORY"], axis=1).head(10)
recent
```

```python
df = recent
df_sample = df[:10]

fig =  ff.create_table(df_sample)
fig.show()
fig.write_image("images/Most_recent_books.jpeg")
```
